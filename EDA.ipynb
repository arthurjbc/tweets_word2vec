```python
import pandas as pd
import seaborn as sns
import re
import json
```

```python
def load_tweets_js(js):
    with open(js, "r", encoding="utf-8") as f:
        content = f.read()
    json_str = re.sub(r"^window\.YTD\.tweets\.part\d+\s*=\s*", "", content.strip())
    data = json.loads(json_str)
    return data

def get_df(js):
    data = load_tweets_js(js)
    tweets = [value["tweet"] for value in data]

    df = pd.json_normalize(tweets)
    df_aux = pd.DataFrame(columns=['text'])
    df_aux['text'] = df['full_text']
    
    return df_aux
     

df = get_df("tweets.js")
display(df.tail(10))
```
